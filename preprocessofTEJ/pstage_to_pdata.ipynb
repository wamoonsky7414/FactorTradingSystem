```python
import pandas as pd

for column in df.columns[1:]:  # Skipping the first column based on your example
    target_dtype = df[column].dtype  # Get the target data type from df
    
    if target_dtype == 'datetime64[ns]':
        if column == '財報發布日':
        # Option 2: Cleaning data by stripping spaces before conversion
            # data['財報發布日'] = data['財報發布日'].str.strip()
            data['財報發布日'] = pd.to_datetime(data['財報發布日'], format='%Y/%m/%d', errors='coerce')
        if column == '財務資料日':
        # Option 2: Cleaning data by stripping spaces before conversion
            # data['財務資料日'] = data['財務資料日'].str.strip()
            data['財務資料日'] = pd.to_datetime(data['財務資料日'], format='%Y%m', errors='coerce')

    elif target_dtype == object:  # Target type is object, which we treat as string here
        # First, attempt to convert to numeric to handle any strings that are actually numbers
        temp_series = pd.to_numeric(data[column], errors='coerce')
        
        # If the series after conversion doesn't become all NaN (meaning it had numeric values)
        if not temp_series.isna().all():
            # We fill NaNs after conversion to avoid issues in converting NaNs to int
            data[column] = temp_series.fillna(0).astype(int).astype(str)
        else:
            # If conversion to numeric fails (resulting in all NaNs), keep as string
            data[column] = data[column].astype(str)
    else:
        # Directly convert to numeric types for everything else, coercing errors
        data[column] = pd.to_numeric(data[column], errors='coerce')
data
```
