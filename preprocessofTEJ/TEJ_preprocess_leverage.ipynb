```python
leverage_data = pd.read_table(rf"{target_folder_path}/籌碼資料庫/融資融券.txt", encoding='Big5-HKSCS', low_memory=False)
leverage_data
```

```python
columns_list = leverage_data.columns.tolist()
the_place_start_to_become_numeric = 4
columns_list
```

```python
leverage_data['年月日'] = pd.to_datetime(leverage_data['年月日'].astype(str).str.strip(), format='%Y%m%d')

string_list = ['公司代碼']
for column_name in string_list:
    leverage_data[column_name] = leverage_data[column_name].str.strip()

numeric_list = [element for element in columns_list[the_place_start_to_become_numeric:] if "發布日" not in element]
for column_name in numeric_list:
    leverage_data[column_name] = leverage_data[column_name].astype(str).str.strip()
    leverage_data[column_name] = pd.to_numeric(leverage_data[column_name], errors='coerce')


leverage_data.rename(columns ={'公司代碼':'symbol', '年月日':'datetime'}, inplace=True)
leverage_data
```

```python
leverage_data.columns = leverage_data.columns.str.replace('/', '')
output_list = [col for col in leverage_data.columns[the_place_start_to_become_numeric:] if col not in 'datetime']
print(len(output_list))
output_list
```

```python
for column in output_list:
    unstack_ohlcv = leverage_data.reset_index().set_index(['datetime','symbol'])[column].unstack()
    csv_file = os.path.join('/Volumes/Database/Factor_data', f'{column}.csv')
    unstack_ohlcv.to_csv(csv_file)
        
```
