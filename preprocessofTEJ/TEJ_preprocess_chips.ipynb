```python
threebig_data = pd.read_table(rf"{target_folder_path}/籌碼資料庫/三大法人買賣超.txt", encoding='Big5-HKSCS', low_memory=False)
threebig_data
```

```python
columns_list = threebig_data.columns.tolist()
the_place_start_to_become_numeric = 4
columns_list
```

```python
threebig_data['年月日'] = pd.to_datetime(threebig_data['年月日'].astype(str).str.strip(), format='%Y%m%d')
# data_full['漲跌停'] =  data_full['漲跌停'].astype(str).str.strip()
# data_full['證券代碼'] = data_full['證券代碼'].str.extract(r'^(\d+)')
string_list = ['證券名稱']
for column_name in string_list:
    threebig_data[column_name] = threebig_data[column_name].str.strip()

numeric_list = [element for element in columns_list[the_place_start_to_become_numeric:] if "發布日" not in element]
for column_name in numeric_list:
    threebig_data[column_name] = threebig_data[column_name].astype(str).str.strip()
    threebig_data[column_name] = pd.to_numeric(threebig_data[column_name], errors='coerce')


threebig_data.rename(columns ={'證券名稱':'symbol', '年月日':'datetime'}, inplace=True)
threebig_data
```

```python
threebig_data.columns = threebig_data.columns.str.replace('/', '')
output_list = [col for col in threebig_data.columns[the_place_start_to_become_numeric:] if col not in 'datetime']
output_list
```

```python
for column in output_list:
    unstack_ohlcv = threebig_data.reset_index().set_index(['datetime','symbol'])[column].unstack()
    csv_file = os.path.join('/Volumes/Database/Factor_data', f'{column}.csv')
    unstack_ohlcv.to_csv(csv_file)
        
```
